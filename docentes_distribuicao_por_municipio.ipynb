```python
import pandas as pd
```

```python
escolas_cariri = pd.read_csv('/content/drive/MyDrive/DataJor/Dados Educação Cariri/escolas_cariri.csv')
escolas_cariri.head()
```

```python
escolas_cariri.info()
```

```python
for col in escolas_cariri.columns:
    if col.startswith('QT_'):
        escolas_cariri[col] = escolas_cariri[col].fillna(0).astype(int)
```

```python
escolas_cariri
```

```python
escolas_cariri.to_csv('escolas_cariri_tratado.csv', index=False)
```

```python
# Seleciona todas as colunas que iniciam com 'QT_DOC' e as colunas com dados dos municípios
colunas_docentes = [col for col in escolas_cariri.columns if col.startswith('QT_DOC')]
colunas_municipio = ['CO_MUNICIPIO', 'NO_MUNICIPIO']
colunas_selecionadas = colunas_municipio + colunas_docentes

# Filtra o DataFrame para conter apenas as colunas selecionadas
escolas_qt_doc = escolas_cariri[colunas_selecionadas]

# Group by municipality and sum the 'QT_DOC' columns
docentes_por_municipio = escolas_qt_doc.groupby(colunas_municipio)[colunas_docentes].sum().reset_index()

# Display the result
display(docentes_por_municipio)
```

```python
docentes_por_municipio.to_csv('docentes_por_municipio.csv', index=False)
```
