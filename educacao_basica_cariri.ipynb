```python
import pandas as pd
```

```python
#IMPORTA A BASE DE DADOS PARA A VARIÁVEL bd_escolas

bd_escolas = pd.read_csv('/content/drive/MyDrive/DataJor/Dados Educação Cariri/microdados_ed_basica_2024.csv', sep=';', encoding='latin-1')
```

```python
bd_escolas
```

```python
#FILTRA AS ESCOLAS DO CEARÁ E A ATRIBUI À VARIÁVEL bd_escolas_ce

bd_escolas_ce = bd_escolas[bd_escolas['SG_UF'] == 'CE']
bd_escolas_ce
```

```python
#SOMA AS MATRÍCULAS E ATRIBUI À VARIÁVEL matriculas_ce     ---->      2.132.934
#DEVE SER LEVADO EM CONSIDERAÇÃO QUE A COLUNA "QT_MAT_BAS" ESTÁ EM FORMATO FLOAT. VALE A CONVERSÃO PARA NÚMERO INTEIRO
#HÁ CÉLULAS VAZIAS E NA. CONVERTER EM "0" ANTES DE CONVERTER

bd_escolas_ce['QT_MAT_BAS'] = bd_escolas_ce['QT_MAT_BAS'].fillna(0)
bd_escolas_ce['QT_MAT_BAS'] = bd_escolas_ce['QT_MAT_BAS'].astype(int)

matriculas_ce = bd_escolas_ce['QT_MAT_BAS'].sum()
matriculas_ce
```

```python
bd_escolas_ce.to_csv('escolas_ce.csv', index=False)
```

```python
#CRIA UMA LISTA COM OS NOMES DOS MUNICÍPIOS DO CARIRI E FILTRA A BASE
municipios_cariri = ['Abaiara', 'Altaneira', 'Antonina do Norte', 'Araripe',
                     'Assaré', 'Aurora', 'Barbalha', 'Barro', 'Brejo Santo',
                     'Campos Sales', 'Caririaçu', 'Crato', 'Farias Brito',
                     'Granjeiro', 'Jardim', 'Jati', 'Juazeiro do Norte',
                     'Lavras da Mangabeira', 'Mauriti', 'Milagres', 'Missão Velha',
                     'Nova Olinda', 'Penaforte', 'Porteiras', 'Potengi', 'Salitre',
                     'Santana do Cariri', 'Tarrafas', 'Várzea Alegre']

bd_escolas_cariri = bd_escolas_ce[bd_escolas_ce['NO_MUNICIPIO'].isin(municipios_cariri)]
bd_escolas_cariri
```

```python
#TOTAL DE MATRÍCULAS NO CARIRI     ------>      254.062

matriculas_cariri = bd_escolas_cariri['QT_MAT_BAS'].sum()
matriculas_cariri
```

```python
bd_escolas_cariri.to_csv('escolas_cariri.csv', index=False)
```

# Task
Analise os dados de educação básica do Cariri, Ceará, contidos no arquivo "/content/escolas_cariri.csv". Calcule o total de matrículas na região (coluna "QT_MAT_BAS"), o total de matrículas por município ("NO_MUNICIPIO"), identifique as escolas ("NO_ENTIDADE") com 0 matrícula e seus respectivos municípios, compare o total de matrículas por município e visualize essa comparação em um gráfico de barras.

## Calcular o total de matrículas no cariri

### Subtask:
Somar os valores da coluna 'QT_MAT_BAS' para obter o total de matrículas na região do Cariri.


**Reasoning**:
Calculate the sum of the 'QT_MAT_BAS' column in the `bd_escolas_cariri` DataFrame, store it in `total_matriculas_cariri`, and print the result to complete the subtask.



```python
total_matriculas_cariri = bd_escolas_cariri['QT_MAT_BAS'].sum()
print(total_matriculas_cariri)
```

## Calcular o total de matrículas por cidade

### Subtask:
Agrupar o DataFrame por 'NO_MUNICIPIO' e somar a coluna 'QT_MAT_BAS' para cada município.


**Reasoning**:
Group the DataFrame by municipality and sum the basic enrollment column to get the total enrollments per municipality.



```python
matriculas_por_municipio = bd_escolas_cariri.groupby('NO_MUNICIPIO')['QT_MAT_BAS'].sum()
matriculas_por_municipio = matriculas_por_municipio.sort_values(ascending=False)
display(matriculas_por_municipio)
matriculas_por_municipio.to_csv('matriculas_por_municipio.csv', index=True)
```

## Identificar escolas com 0 matrícula

### Subtask:
Filtrar o DataFrame para encontrar as escolas onde 'QT_MAT_BAS' é igual a 0 e exibir o nome da escola ('NO_ENTIDADE') e o município ('NO_MUNICIPIO').


**Reasoning**:
Filter the DataFrame to find schools with 0 enrollments and display the school name and municipality.



```python
escolas_sem_matricula = bd_escolas_cariri[bd_escolas_cariri['QT_MAT_BAS'] == 0]
escolas_sem_matricula = escolas_sem_matricula[['NO_ENTIDADE', 'NO_MUNICIPIO']]
display(escolas_sem_matricula)
```

## Comparar o total de matrículas por municípios

### Subtask:
Exibir a tabela com o total de matrículas por município calculada no passo 2 para comparação.


**Reasoning**:
Display the Series containing the total number of enrollments per municipality for comparison as requested in the subtask.



```python
display(matriculas_por_municipio)
```

## Visualizar o total de matrículas por municípios

### Subtask:
Criar um gráfico de barras para visualizar a comparação do total de matrículas entre os municípios.


**Reasoning**:
Create a bar plot to visualize the total enrollments per municipality.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 7))
matriculas_por_municipio.plot(kind='bar', )
plt.title('Total de Matrículas por Município no Cariri')
plt.xlabel('Município')
plt.ylabel('Total de Matrículas')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
